```python
import pandas as pd

# Charger les fichiers CSV
resultats_df = pd.read_csv("scrapping/resultats_enrichis/resultats_enrichis_complets.csv")
crm_df = pd.read_csv("crm_scopa.csv")

# Préparer les colonnes CRM
colonnes_crm = crm_df.columns.tolist()
resultats_formates_df = pd.DataFrame(columns=colonnes_crm)

# Remplir les colonnes avec les données des résultats
resultats_formates_df["Company Name for Emails"] = resultats_df["Nom"]
resultats_formates_df["Email"] = resultats_df["Email trouvé"]
resultats_formates_df["Website"] = resultats_df["Site web"]
resultats_formates_df["Company Address"] = resultats_df["Adresse"]
resultats_formates_df["Statut"] = "Non contacté"

origine_contact = "scrapping python batiment"
resultats_formates_df["origine_contact"] = origine_contact

resultats_formates_df = resultats_formates_df.dropna(subset=["Email"])

emails_crm = crm_df["Email"].dropna().str.lower().unique()
resultats_formates_df = resultats_formates_df[~resultats_formates_df["Email"].str.lower().isin(emails_crm)]

print(f"{len(resultats_formates_df)} contacts {origine_contact} ajoutés")
# Fusionner avec le CRM existant
crm_enrichi_df = pd.concat([crm_df, resultats_formates_df], ignore_index=True)

# Enregistrer dans un nouveau fichier CSV
crm_enrichi_df.to_csv("crm_scopa.csv", index=False)

print("Le fichier enrichi a été enregistré sous 'crm_scopa_enrichi_formaté.csv'")
```
